```python
import pandas as pd
import numpy as np
import random
from catboost import CatBoostRegressor, CatBoostClassifier
from sklearn.metrics import classification_report 
from sklearn.metrics import f1_score
from sklearn.metrics import precision_recall_curve

from sklearn.feature_selection import RFE
from sklearn.model_selection import train_test_split
```

#### Data read

```python
df = pd.read_csv(data_path + 'data/train.csv')
testdf = pd.read_csv(data_path + 'data/test.csv')
```

```python
dt.x.hist()
```

#### Compute aggrs in rolling window

```python
%%time

def percentile(n):
    def percentile_(x):
        return np.percentile(x, n)
    percentile_.__name__ = 'percentile_%s' % n
    return percentile_

feats = dt.x.rolling(5).agg(
    {
     "sum": "sum",
     "min": "min",
     "max": "max",
     "mean": "mean",
     "median": "median",
     "q0.05": percentile(5),
     "q0.25": percentile(25),
     "q0.5": percentile(50),
     "q0.75": percentile(75),
     "q0.95": percentile(95),
  }
)

reverse_feats = dt.x.rolling(5).agg(
    {
     "sum_r": "sum",
     "min_r": "min",
     "max_r": "max",
     "mean_r": "mean",
     "median_r": "median",
     "q0.05_r": percentile(5),
     "q0.25_r": percentile(25),
     "q0.5_r": percentile(50),
     "q0.75_r": percentile(75),
     "q0.95_r": percentile(95),
  }
).shift(-4)

reverse_feats
```

```python
%%time

test_feats = testdf.x.rolling(5).agg(
    {
     "sum": "sum",
     "min": "min",
     "max": "max",
     "mean": "mean",
     "median": "median",
     "q0.05": percentile(5),
     "q0.25": percentile(25),
     "q0.5": percentile(50),
     "q0.75": percentile(75),
     "q0.95": percentile(95),
  }
)

reverse_test_feats = testdf.x.rolling(5).agg(
    {
     "sum_r": "sum",
     "min_r": "min",
     "max_r": "max",
     "mean_r": "mean",
     "median_r": "median",
     "q0.05_r": percentile(5),
     "q0.25_r": percentile(25),
     "q0.5_r": percentile(50),
     "q0.75_r": percentile(75),
     "q0.95_r": percentile(95),
  }
).shift(-4)

reverse_test_feats
```

```python
df.shape, feats.shape, test_feats.shape
```

#### Merge dfs

```python
dt = df.merge(feats, left_index=True, right_index=True, how="left")
dt = dt.merge(reverse_feats, left_index=True, right_index=True, how="left")

testdt = testdf.merge(test_feats, left_index=True, right_index=True, how="left")
testdt = testdt.merge(reverse_test_feats, left_index=True, right_index=True, how="left")


print(dt.shape, testdt.shape)
testdt
```

```python
cols = ["x", "sum", "min", "max", "mean", "median", "q0.05", "q0.25", "q0.5", "q0.75", "q0.95",
  "sum_r", "min_r", "max_r", "mean_r", "median_r", "q0.05_r", "q0.25_r", "q0.5_r", "q0.75_r", "q0.95_r"]
```

```python
import matplotlib.pyplot as plt

for col in cols:
#     dt[(dt[col] < dt[col].quantile(0.95)) &
#        (~dt['crash_type'].isin(dt['crash_type'].unique()[:-1]))][col].hist(density=True)

    dt[col].hist(density=True)
    testdt[col].hist(density=True)
    plt.title(col)
    plt.show()
```

#### Split data

```python
ids = pd.unique(dt.id)
```

```python
# train = dt[dt['id'].isin(random.sample(range(0, len(ids)), round(len(ids)*.8)))]
# test = dt[~dt['id'].isin(random.sample(range(0, len(ids)), round(len(ids)*.8)))]

ids_train, ids_test = train_test_split(ids, test_size=0.15, random_state=42)

train = dt[dt.id.isin(ids_train)]
test = dt[dt.id.isin(ids_test)]
train.shape, test.shape
```

```python
train.drop(["id"], inplace=True, axis=1)
test.drop(["id"], inplace=True, axis=1)
```

```python
cols
```

```python
X_test = test[cols]
y_test = test['y']
X = train[cols]
y = train['y']

X_sub = testdt[cols]
X, X_test, X_sub
```

```python
print(X.isna().any())
print(X_test.isna().any())
X.fillna(0, inplace=True)
X_test.fillna(0, inplace=True)
X_sub.fillna(0, inplace=True)
```
